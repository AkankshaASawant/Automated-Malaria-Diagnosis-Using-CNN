## Build the VGG-19 Architecture

```python
# Define size of image
image_size = [224,224]

# Import Vgg16 library as shown and add preprocessing layer in front og vgg
vgg = VGG19(input_shape=image_size + [3], weights='imagenet', include_top=False) # Weights:- Imagenet data
```

```python
# Dont train the existing weights
for layer in vgg.layers:
    layer.trainable = False
```

```python
# Assuming `vgg` is your VGG19 base model
x = Flatten()(vgg.output)
x = Dropout(0.5)(x)
x = Dense(1024, activation='relu')(x)  # Added hidden dense layer
x = Dropout(0.5)(x)
x = Dense(512, activation='relu')(x)  # Added hidden dense layer
x = Dropout(0.5)(x)
x = Dense(256, activation='relu')(x)  # Added hidden dense layer
x = Dropout(0.5)(x)
prediction = Dense(2, activation='sigmoid')(x)
vgg19_model = Model(inputs=vgg.input, outputs=prediction)
```

## Simple SequeSequential Model

```python
from keras import models, layers


model = models.Sequential()
model.add(layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu', input_shape=(224,224,3)))
model.add(layers.BatchNormalization())
model.add(layers.MaxPool2D(pool_size=(2,2)))
model.add(layers.Dropout(rate=0.3))

model.add(layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPool2D(pool_size=(2,2)))
model.add(layers.Dropout(rate=0.3))

model.add(layers.Conv2D(filters=128, kernel_size=(3,3), activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPool2D(pool_size=(2,2)))
model.add(layers.Dropout(rate=0.4))

model.add(layers.Conv2D(filters=256, kernel_size=(3,3), activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPool2D(pool_size=(2,2)))
model.add(layers.Dropout(rate=0.4))

model.add(layers.Conv2D(filters=512, kernel_size=(3,3), activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPool2D(pool_size=(2,2)))
model.add(layers.Dropout(rate=0.5))

model.add(layers.Flatten())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dropout(rate=0.5))
model.add(layers.Dense(2, activation='sigmoid'))
```

## RETRAINED SEQUENTIAL MODEL

```python
new_model = models.Sequential()

# First Conv Block
new_model.add(layers.Conv2D(filters=32, kernel_size=(3,3), activation=None, input_shape=(224, 224, 3)))
new_model.add(layers.BatchNormalization())
new_model.add(layers.Activation('relu'))
new_model.add(layers.MaxPool2D(pool_size=(2,2)))
new_model.add(layers.Dropout(rate=0.3))

# Second Conv Block
new_model.add(layers.Conv2D(filters=64, kernel_size=(3,3), activation=None, kernel_regularizer=regularizers.l2(0.01)))
new_model.add(layers.BatchNormalization())
new_model.add(layers.Activation('relu'))
new_model.add(layers.MaxPool2D(pool_size=(2,2)))
new_model.add(layers.Dropout(rate=0.3))

# Third Conv Block
new_model.add(layers.Conv2D(filters=128, kernel_size=(3,3), activation=None, kernel_regularizer=regularizers.l2(0.01)))
new_model.add(layers.BatchNormalization())
new_model.add(layers.Activation('relu'))
new_model.add(layers.MaxPool2D(pool_size=(2,2)))
new_model.add(layers.Dropout(rate=0.4))

# Fourth Conv Block
new_model.add(layers.Conv2D(filters=256, kernel_size=(3,3), activation=None, kernel_regularizer=regularizers.l2(0.01)))
new_model.add(layers.BatchNormalization())
new_model.add(layers.Activation('relu'))
new_model.add(layers.MaxPool2D(pool_size=(2,2)))
new_model.add(layers.Dropout(rate=0.4))

# Fully Connected Block
new_model.add(layers.Flatten())
new_model.add(layers.Dense(256, activation=None, kernel_regularizer=regularizers.l2(0.01)))
new_model.add(layers.BatchNormalization())
new_model.add(layers.Activation('relu'))
new_model.add(layers.Dropout(rate=0.5))
new_model.add(layers.Dense(2, activation='sigmoid'))
```

### ENSEMBLE APPORCH

```python
# Load your trained models
vgg19_model = load_model('VGG19_model.keras')
vgg16_model = load_model('NEW_VGG16_model.keras')
dcnn_model = load_model('new_seq_model.keras')

# Ensemble models by averaging their predictions
def ensemble_models(models, model_input):
    outputs = [model.predict(model_input, verbose=0) for model in models]
    y = np.array(outputs)
    avg = np.mean(y, axis=0)
    return avg

# Assuming you have your test data loaded in X_test
ensemble_predictions = ensemble_models([vgg19_model, vgg16_model, dcnn_model], X_test)

# Convert ensemble predictions to class labels
final_predictions_binary = np.argmax(ensemble_predictions, axis=1)
y_test_binary = np.argmax(y_test, axis=1) if y_test.shape[1] > 1 else y_test
```

```python
# Calculate accuracy, F1 score, and classification report
accuracy = accuracy_score(y_test_binary, final_predictions_binary)
f1 = f1_score(y_test_binary, final_predictions_binary, average='weighted')
report = classification_report(y_test_binary, final_predictions_binary)

print(f'Ensemble Model Accuracy: {accuracy}')
print(f"F1 Score: {f1}")
print(f"Classification Report:\n{report}")
```

## Testing Random Images

```python
import gradio as gr
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array,load_img
```

```python
# Load your model
model = load_model("NEW_VGG16_model.keras")

# Define the class names
class_names = ['Infected','Uninfected']


# Preprocess the image
def preprocess_image(image):
    image = cv2.resize(image, (224,224))
    image = img_to_array(image)
    image = image / 255.0  # Scale pixel values
    image = tf.expand_dims(image, axis=0)  # Add batch dimension
    return image

# Create function for prediction
def predict(image):
    image_np = np.array(image) # Convert PIL Image to numpy array
    processed_image = preprocess_image(image_np) # Preprocess the CLAHE image
    prediction = model.predict(processed_image)  # Make prediction
    class_idx = tf.argmax(prediction, axis=1).numpy()[0]
    confidence = tf.reduce_max(prediction).numpy()
    return class_names[class_idx], f"{confidence:.2f}"
```

```python
# Create the Gradio interface
iface = gr.Interface(
    fn=predict,
    inputs=gr.Image(type='pil', label='Upload an Image'),
    outputs=[gr.Label(label='Predicted Class'),
             gr.Textbox(label='Confidence')],
    title='Malaria Cell Classification',
    description='Upload an cell image to classify it as Infected and Uninfected cell.')

# Launch the interface
iface.launch(share=True)
```
